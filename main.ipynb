```
import os, textwrap
from moviepy.editor import TextClip, CompositeVideoClip, ColorClip
import pysrt

# Config
import json
with open('config.json') as f:
    config = json.load(f)

# Choose input method
USE_TEXT_FILE = True  # Set False to paste text

if USE_TEXT_FILE:
    with open('script.txt','r') as f:
        script = f.read().strip()
else:
    script = """Paste your script here..."""

# Split into sentences
sentences = [s.strip() for s in script.replace('\n',' ').split('.') if s.strip()]

# Generate SRT
subs = pysrt.SubRipFile()
start_sec = 0
for i, sentence in enumerate(sentences,1):
    end_sec = start_sec + config['seconds_per_sentence']
    subs.append(pysrt.SubRipItem(
        index=i,
        start=pysrt.SubRipTime(seconds=start_sec),
        end=pysrt.SubRipTime(seconds=end_sec),
        text=sentence
    ))
    start_sec = end_sec
subs.save('subtitles.srt', encoding='utf-8')

print('✅ subtitles.srt generated')
```
